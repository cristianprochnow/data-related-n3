# Contexto

Aqui o foco é justamente compactar os dados em um formato performático para o restante da análise, que é o `.parquet`. Esse formato garante que os dados possam ser interpretados sem dificuldades no restante do processo e da melhor forma possível, evitando ambiguidades.

```python
import pandas as pd

raw_path = '../data/salaries.csv'
bronze_path = '../data/bronze/salaries.parquet'

df_raw = pd.read_csv(raw_path)

print(f"Linhas carregadas: {df_raw.shape[0]}")
display(df_raw.head())

df_raw.to_parquet(bronze_path, index=False)

print(f"Dados salvos com sucesso na camada Bronze: {bronze_path}")
```
